**Question 6**

**Question 7**

Which state recorded the most delays in the 2nd Quarter of 2019

```python
#Calculate Total delay as sum of Arrival+Departure Delay
Second_Quater['TOT_DELAY']=Second_Quater['ARR_DELAY']+Second_Quater['DEP_DELAY']
```

```python
#Gets the Unique Origin State
Sec_Quat=Second_Quater['ORIGIN_STATE_NM'].unique()
Sec_Qua=pd.DataFrame(Sec_Quat)
```

```python
Sec_Qua.columns=['ORIGIN_STATE']
```

```python
Sec_Qua.head()
```

```python
#Calculate the Total delay for each Unique Origin State
for i in Sec_Quat:
     
    j=(Second_Quater['TOT_DELAY'][Second_Quater['ORIGIN_STATE_NM']==i]).sum()
        
    print(j)
```

```python
#The results are saved as a csv which I read again
TOTAL=pd.read_csv('TOT_DELAY.csv')
TOTAL.head()
```

```python
FIN=pd.concat([Sec_Qua,TOTAL],axis=1)
```

```python
FIN['TOT_STATE_DEL'].max()
```

```python
#Locates the State with the maximum Total delay
FIN.loc[FIN['TOT_STATE_DEL']==5857487]
```

From the results Texas has the highest total delays of 5857487

**Question 8**

What is the daily average Taxi-out time in New York in February 2019 

```python
Feb.head()
```

```python
Feb2=pd.DataFrame(Feb, columns=['DAY_OF_MONTH','ORIGIN_STATE_NM','TAXI_OUT'])
```

```python
Feb2.head()
```

```python
Feb2['ORIGIN_STATE_NM'].unique()
```

```python
#Gets the data for new york
Feb2=Feb2[Feb2['ORIGIN_STATE_NM']=='New York']
```

```python
#Finds  the average for each day departure
FEB2=Feb2.groupby('DAY_OF_MONTH').mean().head()
```

```python
#Finds the average Taxi-Out Time for each
FEB2['TAXI_OUT'].mean()
```

The Daily Average Taxi out Time in New york is 19.25
